```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings


from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import scipy.cluster.hierarchy as sch

warnings.filterwarnings('ignore', category=UserWarning)
warnings.filterwarnings('ignore', category=FutureWarning)


sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

print("Environment verified. Libraries loaded successfully.")
```

```python

print("Reading target data assets from local repository...")

df_clients = pd.read_csv('clients.csv')
df_properties = pd.read_csv('properties.csv')

print(f"-> Verification Success: Loaded {len(df_clients)} client profiles.")
print(f"-> Verification Success: Loaded {len(df_properties)} total property listings.")
```

```python

df_sold_units = df_properties[df_properties['listing_status'].astype(str).str.strip() == 'Sold'].copy()


df_sold_units['sale_price'] = (
    df_sold_units['sale_price']
    .astype(str)
    .str.replace('$', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.replace('"', '', regex=False)
    .str.strip()
    .astype(float)
)

print(f"Isolated {len(df_sold_units)} validated transaction rows. Revenue metrics parsed successfully.")
```

```python

client_portfolios = df_sold_units.groupby('client_ref').agg(
    total_investment=('sale_price', 'sum'),
    units_purchased=('listing_id', 'count')
).reset_index()

print("Portfolio matrices generated per transacting buyer identifier:")
print(client_portfolios.head())
```

```python

df_master = pd.merge(
    df_clients, 
    client_portfolios, 
    left_on='client_id', 
    right_on='client_ref', 
    how='left'
)

df_master['total_investment'] = df_master['total_investment'].fillna(0.0)
df_master['units_purchased'] = df_master['units_purchased'].fillna(0).astype(int)

print(f"Master consolidation structural matrix verified. Shape: {df_master.shape}")
df_master.head()
```

```python

df_cleaned = df_master.copy()


initial_rows = len(df_cleaned)
df_cleaned.drop_duplicates(subset=['client_id'], keep='first', inplace=True)
print(f"Deduplication complete: Eliminated {initial_rows - len(df_cleaned)} rows.")


if df_cleaned['satisfaction_score'].isnull().sum() > 0:
    median_score = df_cleaned['satisfaction_score'].median()
    df_cleaned['satisfaction_score'] = df_cleaned['satisfaction_score'].fillna(median_score)
    print(f"Imputed missing survey fields with dataset median value: {median_score}")

categorical_cols = ['client_type', 'gender', 'country', 'region', 'acquisition_purpose', 'loan_applied', 'referral_channel']
for col in categorical_cols:
    df_cleaned[col] = df_cleaned[col].astype(str).str.strip()

print("Data cleaning step completed successfully.")
```

```python

df_cleaned['date_of_birth'] = pd.to_datetime(df_cleaned['date_of_birth'], errors='coerce')


if df_cleaned['date_of_birth'].isnull().sum() > 0:
    fallback_dates = pd.to_datetime(df_master['date_of_birth'], format='mixed', errors='coerce')
    df_cleaned['date_of_birth'] = df_cleaned['date_of_birth'].fillna(fallback_dates)


target_anchor_year = 2026 
df_cleaned['Age'] = target_anchor_year - df_cleaned['date_of_birth'].dt.year


df_cleaned['Age'] = df_cleaned['Age'].fillna(df_cleaned['Age'].median()).astype(int)


numerical_features = ['Age', 'satisfaction_score', 'total_investment', 'units_purchased']
categorical_features = ['client_type', 'gender', 'country', 'region', 'acquisition_purpose', 'loan_applied', 'referral_channel']


X_numerical = df_cleaned[numerical_features].copy()
X_categorical = pd.get_dummies(df_cleaned[categorical_features], drop_first=True)


X_combined = pd.concat([X_numerical, X_categorical], axis=1)


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_combined)

print(f"Matrix features processed. Engineering feature array shape: {X_scaled.shape}")
```

```python

wcss = []
silhouette_scores = []
k_range = range(2, 11)

for k in k_range:
    kmeans_engine = KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10)
    cluster_labels = kmeans_engine.fit_predict(X_scaled)
    
    wcss.append(kmeans_engine.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, cluster_labels))


fig, ax1 = plt.subplots(figsize=(12, 5))

color_metric = 'tab:blue'
ax1.set_xlabel('Number of Target Clusters (k)')
ax1.set_ylabel('Inertia (Within-Cluster Sum of Squares)', color=color_metric)
ax1.plot(k_range, wcss, marker='o', color=color_metric, linewidth=2, label='Inertia')
ax1.tick_params(axis='y', labelcolor=color_metric)

ax2 = ax1.twinx()
color_score = 'tab:orange'
ax2.set_ylabel('Silhouette Score Coefficient', color=color_score)
ax2.plot(k_range, silhouette_scores, marker='s', color=color_score, linewidth=2, linestyle='--', label='Silhouette')
ax2.tick_params(axis='y', labelcolor=color_score)

plt.title('Clustering Optimization Diagnostics: Locating the Structural Elongation Point')
fig.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(14, 6))

# Execute mathematical tree generation via Ward's variance minimization technique
linkage_tree = sch.linkage(X_scaled, method='ward')
dendrogram_renderer = sch.dendrogram(linkage_tree, no_labels=True)

plt.title('Hierarchical Structure Analysis (Dendrogram Validation Plot)')
plt.xlabel('Customer Profile Observation Matrix Row Indices')
plt.ylabel('Variance Distance Linkage Cut-off Threshold')
plt.show()
```

```python

target_clusters = 4
final_segmentation_model = KMeans(n_clusters=target_clusters, init='k-means++', random_state=42, n_init=10)
df_cleaned['Cluster_ID'] = final_segmentation_model.fit_predict(X_scaled)


print("--- SUMMARY MEAN METRICS PER RAW K-MEANS ASSIGNED NUMBER ---")
summary_profiles = df_cleaned.groupby('Cluster_ID')[numerical_features].mean()
print(summary_profiles)

cluster_code_map = {0: 'C1', 1: 'C2', 2: 'C3', 3: 'C4'}
buyer_persona_names = {
    'C1': 'Global Investors',
    'C2': 'First-Time Buyers',
    'C3': 'Corporate Buyers',
    'C4': 'Luxury Investors'
}

df_cleaned['Cluster_Code'] = df_cleaned['Cluster_ID'].map(cluster_code_map)
df_cleaned['Buyer_Segment'] = df_cleaned['Cluster_Code'].map(buyer_persona_names)


df_cleaned.to_csv('segmented_market_intelligence.csv', index=False)
print("\nSuccess! Combined analysis engine dataset saved to 'segmented_market_intelligence.csv'.")


fig, axes = plt.subplots(1, 2, figsize=(16, 6))
persona_plot_order = ['Global Investors', 'First-Time Buyers', 'Corporate Buyers', 'Luxury Investors']

sns.barplot(data=df_cleaned, x='Buyer_Segment', y='total_investment', estimator=np.mean, palette='Blues_r', ax=axes[0], order=persona_plot_order)
axes[0].set_title('Average Total Capital Expended per Market Segment ($)')
axes[0].tick_params(axis='x', rotation=15)

sns.boxplot(data=df_cleaned, x='Buyer_Segment', y='Age', hue='loan_applied', palette='Set2', ax=axes[1], order=persona_plot_order)
axes[1].set_title('Demographic Age Dispersions Grouped by Financing Requests')
axes[1].tick_params(axis='x', rotation=15)

plt.tight_layout()
plt.show()
```
